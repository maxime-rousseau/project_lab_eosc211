```python
def grav_acc(x, y,  Mp = 5.9722*10**24):
    """
    Calculate the gravitational acceleration at a point (x, y) due to a mass M located at the origin.

    Parameters:
    x (float): x-coordinate of the point in km
    y (float): y-coordinate of the point in km
    M (float): mass located at the origin

    Returns:
    tuple: gravitational acceleration components (ax, ay) units in m/s^2
    """
    G = 6.67430e-11  # gravitational constant in m^3 kg^-1 s^-2
    x *= 1000  # convert km to m
    y *= 1000  # convert km to m
    r = (x**2 + y**2)**0.5
    a = G * Mp / r**2
    
    ax = -a * (x / r)
    ay = -a * (y / r)

    return ax, ay

# Example usage:
ax, ay = grav_acc(0, 6371)  # Gravitational acceleration at Earth's surface
print(f"Gravitational acceleration at (0, 6371 km): ax = {ax:.2e} m/s^2, ay = {ay:.2e} m/s^2")
#yes it is my expected output

ax, ay = grav_acc(-4787, -4787)
print(f"Gravitational acceleration at (-4787 km, -4787 km): ax = {ax:.2e} m/s^2, ay = {ay:.2e} m/s^2")
# it makes sense for ax ay absolute values to be equal since the point is equidistant from both axes, 
# the acceleration components should be equal in magnitude.

ax, ay = grav_acc(6771, 0)
print(f"Gravitational acceleration at (6771 km, 0 km): ax = {ax:.2e} m/s^2, ay = {ay:.2e} m/s^2")
# it makes sense for ay to be 0 since the point lies on the x-axis,
# so there should be no vertical component of gravitational acceleration.
```
