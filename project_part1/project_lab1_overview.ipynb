Read the background reading BEFORE attempting the lab.
The background reading is contained in the file `project_lab1_reading.ipynb` and includes learning goals.

For this project lab, you will write 4 functions as follows:



**In Part 1** you will write the function `grav_acc` which calculates the gravitational acceleration on the spacecraft at any given point along the trajectory and is at the heart of all the calculations. You will check that it runs successfully by calling it for 3 test cases for which we give you the result.  You will also make a "top-down" plan for coding up a full fly-by simulation by writing pseudo-code to plan out your algorithm.

**In Part 2** you will write the remaining functions `checkinit`, `sc_vel_pos_change` and `get_traj`. You will make a figure to verify that you can reproduce a spacecraft trajectory provided. 
- `get_traj` computes the acceleration, velocity and position vectors along a spacecraft's trajectory and will call three subfunctions: `checkinit`, `grav_acc`, and `sc_vel_pos_change`.

- `checkinit` checks that your initial spacecraft velocity and position are feasible.

**In Part 3** you will write additional script(s) to call `get_traj` for various choices of parameters for starting position / velocity, total duration of flyby and the time step. The script(s) will contain the code to generate the requested figures or output for each investigation.

## 4. Important Coding Points:
1. **The use of SI units is ESSENTIAL**. Make sure to convert ALL your length, mass and time variables into m, kg and seconds before doing any calculations. For plotting purposes it is more useful to use km, so convert m to km only for plotting.
2. **Don't round the output of any calculations, ESPECIALLY inside functions.** Rounding errors propagate and can result in final answers that are quite different (and potentially catastrophic - e.g. your spacecraft crashes when you thought it didn't ...).  Obviously when printing output files or labeling figures rounding calculations to a sensible choice of decimal places (where "sensible choice" depends on the situation) is desirable.
2. The amount of code required is not huge. For example, my version of `get_traj` contains only 13 lines of code and the subfunctions `grav_acc`, `sc_vel_pos_change`, and `checkinit` contain 5, 3 and 3 lines of code respectively, not including comment and help lines. However, you should be vigilent about being consistent in your use of variable names, and in not generating unecessary variables, otherwise your code (and thinking!) will get out of hand.
3. Make sure your code is organized into easily readable, compact coding “blocks”, with appropriate comment lines, and separated by white space.
4. All functions must contain docstrings.
5. Suppress any extra `print` statements in your notebook that aren't necessary.
