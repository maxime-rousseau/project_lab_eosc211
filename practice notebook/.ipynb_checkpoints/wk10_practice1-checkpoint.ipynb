# Practice with positional and keyword arguments

The function below computes the gravity anomaly in units known as milligals (mgal) that results from topography with specified density. The equation shows that 1000 m of topograhy with a density of 1000 kg/m^3, i.e. 1 km of ice, gives a gravity anomay ~42 mgal. Similarly negative topography gives a negative gravity anomaly.  So a hole 1 km deep in ice gives an anomaly of -42 mgal because the ice mass is essentially missing in the hole.

**What is printed to the screen after running the code in each of 1a - 1g?**

```python
import numpy as np

def grav_anom(topo, dens=1000):
    """
    Computes gravity anomaly that results from topography with specified density.
    Equation showsthat 1 km (1000m) of topo with a density 
       of 1000 kg/m^3 gives a gravity anomay ~ 42 mgal 
    
    Inputs: topography in meters, density in kg/m^3
    
    Outputs:  gravity anomaly in mgal
    
    """
    
    grav = 42*(dens/1000)*(topo/1000)
    
    return grav
```


#### 1a)
```python 
g1 = grav_anom(1000,1000)
print(f'{g1 = }')  
```
<br><br>

#### 1b)
```python
g2 = grav_anom(1000)
print(f'{g2 = }')  
```
